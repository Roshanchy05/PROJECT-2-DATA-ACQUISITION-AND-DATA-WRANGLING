```python
import pandas as pd
```

```python
df1 = pd.read_csv("dataset_1 - dataset_1.csv"))
```

```python
df1.head()
```

```python
df2 = pd.read_csv("dataset_2.xlsx - dataset_2.csv")
```

```python
df2.head()
```

```python
try:
    df3 = pd.read_csv("dataset_3 - dataset_3.csv", encoding='utf-8')
except:
    df3 = pd.read_csv("dataset_3 - dataset_3.csv", encoding='latin1')
```

```python
df3.head()
```

```python
merged_df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
print(merged_df.head())
```

```python
for column in merged_df.columns:
    print(f"Unique values in '{column}': {merged_df[column].unique()}\n")
```

```python
merged_df = merged_df.loc[:, ~merged_df.columns.str.contains('^Unnamed')]
```

```python
print(merged_df.columns)
```

```python
print("Shape of the dataset:", merged_df.shape)
```

```python
print("Data types of each column:\n", merged_df.dtypes)
```

```python
merged_df.describe(include='all')
```

```python
print("Missing values in each column:\n", merged_df.isnull().sum())
```

```python
if 'casual' in merged_df.columns and 'registered' in merged_df.columns and 'cnt' in merged_df.columns:
    merged_df['cnt_check'] = merged_df['casual'] + merged_df['registered']
    incorrect_rows = merged_df[merged_df['cnt'] != merged_df['cnt_check']]
    print(f"Number of incorrect rows where cnt != casual + registered: {len(incorrect_rows)}")
    merged_df = merged_df.drop(columns=['cnt_check'])
```

```python
print("Number of duplicate rows:", merged_df.duplicated().sum())
merged_df = merged_df.drop_duplicates()
```

```python
merged_df.to_csv("cleaned_merged_dataset.csv", index=False)
print("Cleaned merged dataset saved as 'cleaned_merged_dataset.csv'")
```

# CHECK SKEWNESS

```python
import numpy as np

skew_values = merged_df.skew(numeric_only=True)
print("Skewness:\n", skew_values)

highly_skewed = skew_values[(skew_values > 1) | (skew_values < -1)]
print("\nHighly skewed columns:\n", highly_skewed)

for col in highly_skewed.index:
    if (merged_df[col] >= 0).all():  # ensure no negative values
        merged_df[col] = np.log1p(merged_df[col])
        print(f"Applied log1p transform to column: {col}")
    else:
        print(f"Skipped column {col} because it contains negative values.")

new_skew_values = merged_df.skew(numeric_only=True)
print("\nNew skewness after transformation:\n", new_skew_values)
```

```python
print("Skewness of numeric columns:\n")
print(merged_df.skew(numeric_only=True))
```
